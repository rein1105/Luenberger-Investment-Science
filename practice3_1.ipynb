```python
import numpy as np
import numpy_financial as npf
```

```python
cashflows = np.array([0] + [8]*17 + [108])
npv = npf.npv(0.09, cashflows)
print(npv)
```

```python
def cashflows_bond(face_value, coupon_rate, maturity):
    return np.array([0] + [face_value * coupon_rate] * (maturity - 1) + [face_value * (1 + coupon_rate)])

cashflows = cashflows_bond(100, 0.08, 18)
npv = npf.npv(0.09, cashflows)
print(npv)
```

```python
rate = 0.1
coupon_rate = 0.08
cashflows = cashflows_bond(100, coupon_rate, 10)
price = npf.npv(rate, cashflows)
print(price)
t = np.arange(len(cashflows))
pv_each = cashflows / (1 + rate) ** t
duration = np.sum(t * pv_each / price)
print(duration)
```

```python
rate = 0.15
cashflows_A = cashflows_bond(1000, 0.1, 3)
cashflows_B = cashflows_bond(1000, 0.05, 3)
cashflows_C = cashflows_bond(1000, 0.00, 3)
cashflows_D = cashflows_bond(1000, 0.0, 1)
price_A = npf.npv(rate, cashflows_A)
price_B = npf.npv(rate, cashflows_B)
price_C = npf.npv(rate, cashflows_C)
price_D = npf.npv(rate, cashflows_D)
print(price_A, price_B, price_C, price_D)
def duration(cashflows, price, rate):
    t = np.arange(len(cashflows))
    pv_each = cashflows / (1 + rate) ** t
    return np.sum(t * pv_each) / price
duration_A = duration(cashflows_A, price_A, rate)
duration_B = duration(cashflows_B, price_B, rate)
duration_C = duration(cashflows_C, price_C, rate)
duration_D = duration(cashflows_D, price_D, rate)
print(duration_A, duration_B, duration_C, duration_D)
sensitivity_each = {"A": price_A * duration_A, "B": price_B * duration_B, "C": price_C * duration_C, "D": price_D * duration_D}
most_sensitive = max(sensitivity_each, key=sensitivity_each.get)
print(most_sensitive)
```
