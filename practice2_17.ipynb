```python
import numpy as np
import numpy_financial as npf
import matplotlib.pyplot as plt
```

```python
initial_investment = 10_000_000
tax_rate = 0.34
years = 5
r = 0.12
inflation_rate = 0.04

annual_labor_hours = 10_000
annual_material_tons = 100
wage_per_hour = 30
material_cost_per_ton = 100

annual_units_sold = 1_000_000
price_per_unit = 3.3

annual_dep = initial_investment / years
annual_revenue = annual_units_sold * price_per_unit
annual_operating_cost = annual_labor_hours * wage_per_hour + annual_material_tons * material_cost_per_ton
ebit = annual_revenue - annual_operating_cost - annual_dep

tax = ebit * tax_rate
cashflow = ebit - tax + annual_dep

cashflows = []
for n in range(years + 1):
    if n == 0:
        cashflows.append(-initial_investment)
    else:
        cashflows.append(cashflow)
cashflows = np.array(cashflows)

npv_before = npf.npv(0.08, cashflows)

print(npv_before)#インフレ適用前

#収入と創業費用にインフレ適用後を考える
annual_revenue_after = annual_revenue * (1+inflation_rate)**np.arange(years+1)
annual_operating_cost_after = annual_operating_cost * (1+inflation_rate)**np.arange(years+1)
print(annual_revenue_after)
print(annual_operating_cost_after)
ebit_after = annual_revenue_after - annual_operating_cost_after - annual_dep
tax_after = ebit_after * tax_rate
cashflows_after = ebit_after - tax_after + annual_dep
cashflows_after[0] = -initial_investment
print(cashflows_after)
npv_after = npf.npv(r, cashflows_after)
print(npv_after)
```

```python
t = np.arange(1, years+1)

annual_revenue_after = annual_revenue * (1+inflation_rate)**t
annual_operating_cost_after = annual_operating_cost * (1+inflation_rate)**t

ebit_after = annual_revenue_after - annual_operating_cost_after - annual_dep
tax_after = ebit_after * tax_rate
cashflows_after = np.concatenate(
    ([-initial_investment],
     ebit_after - tax_after + annual_dep)
)

print(cashflows_after)
npv_after = npf.npv(r, cashflows_after)
print(npv_after)
print(npv_after-npv_before)
```
